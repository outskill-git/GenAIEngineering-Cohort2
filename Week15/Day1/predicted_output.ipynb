```python
messages= [{ "role": "system", "content": "You are a helpful assistant." }, 
           { "role": "user", "content": "What is the capital of France?" }]

print(messages)
```

```python
# CHat completion openai call to get a response for messages

import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))


#Chat Completion
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    max_tokens=50,
    prediction={
        "type": "content",
        "content": "The capital of France is Paris." # This could be generated by a cheaper LLM
    }
)


print(response)
print(response.usage)
print("================")
print("Total tokens:")
print(response.usage.total_tokens)
print("Prompt tokens: (Input Tokens)")
print(response.usage.prompt_tokens)
print("Cached Prompt Tokens:")
print(response.usage.prompt_tokens_details.cached_tokens)
print("Completion Tokens: (Output Tokens)")
print("================")
print("Accepted Prediction Tokens:  ", response.usage.completion_tokens_details.accepted_prediction_tokens)
print("Rejected Prediction Tokens:  ", response.usage.completion_tokens_details.rejected_prediction_tokens)
print("================")
print(response.usage.completion_tokens)
print(response.choices[0].message.content)
```
